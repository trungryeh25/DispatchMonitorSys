```python
import torch
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
import os
```

```python
# ====================== 2. Config ======================
# ⚠️ Chọn 'dish' hoặc 'tray'
CLASS_TYPE = 'tray'

DATA_DIR = f'/content/drive/MyDrive/TestExam-preInterview/AU/Dataset/Classification/{CLASS_TYPE}'
NUM_CLASSES = 3
BATCH_SIZE = 32
EPOCHS = 20
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# ====================== 3. Dataset ======================
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

train_dataset = datasets.ImageFolder(DATA_DIR, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
# ====================== 4. Model ======================
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, NUM_CLASSES)
model = model.to(DEVICE)
```

### dish

```python
# ====================== 5. Training ======================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    print(f"Epoch {epoch+1} Loss: {running_loss/len(train_loader):.4f}")
```

```python
# ====================== 6. Save Model ======================
MODEL_PATH = f'/content/drive/MyDrive/TestExam-preInterview/DispatchMonitor/resnet/weights/resnet_{CLASS_TYPE}.pt'
torch.save(model.state_dict(), MODEL_PATH)
print(f"✅ Saved to {MODEL_PATH}")
```
